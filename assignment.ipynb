```python
import seaborn as sns
import pandas as pd

%matplotlib inline
```

```python
df = pd.read_csv('50ulke.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
sns.kdeplot(df['National Income'], shade=True, color='g')
```

```python
sns.distplot(df['Growth'], kde=False, bins=30, color='g')
```

```python
sns.kdeplot(df['Growth'], shade=True, color='g')
```

```python
sns.jointplot(df['National Income'],df['Per person'], data=df, color='g')
```

```python
sns.jointplot(df.Growth,df['Per person'], data=df, color='g')
```

```python
sns.countplot(df.Continent, data=df, palette='viridis')
```

```python
sns.stripplot(df['Continent'],df.Growth, data=df, dodge=True)
```

```python
sns.boxplot(df['Continent'], df.Growth, data=df)
```

```python
sns.violinplot(df['Continent'], df.Growth, data=df)
```

```python
sns.barplot(df['Continent'], df.Growth, data=df)
```
